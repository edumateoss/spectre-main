```python
import albumentations as A
from albumentations.pytorch import ToTensorV2
from pathlib import Path
from torch.utils.data import DataLoader
from spectre.datasets import RFUAVDataset
import torch
from torchvision.models import resnet18

transform = A.Compose([
    A.Resize(224, 224),
    A.Normalize(),
    ToTensorV2(),
])

ds = RFUAVDataset(root=Path("/datasets/rfuav"), transform=transform)
loader = DataLoader(ds, batch_size=4, shuffle=True)

images, targets = next(iter(loader))

model = resnet18(weights="DEFAULT")
model.fc = torch.nn.Linear(model.fc.in_features, len(ds.classes))

outputs = model(images)

print("Shape imágenes:", images.shape)
print("Shape salida:", outputs.shape)
print("Predicciones:", outputs.argmax(dim=1))
print("Reales:", targets)
print("Clases:", ds.classes)
```
